```python
import pandas as pd
import os
import glob

# Ruta a la carpeta
path = 'LinkedIn/Data'  

# Construye la ruta completa y busca todos los archivos CSV en la carpeta
all_files = glob.glob(os.path.join(path, "*.csv"))

# Lista para almacenar cada DataFrame leído de cada archivo CSV
dfs = []

# Itera sobre los archivos encontrados y los añade a la lista de DataFrames
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    dfs.append(df)

# Combina todos los DataFrames de la lista en uno solo
combined_df = pd.concat(dfs, axis=0, ignore_index=True)

# Muestra el DataFrame combinado
print(combined_df)
#combined_df.to_excel("_output/desa_linked_df.xlsx")
```

```python
master_df = combined_df.drop('Unnamed: 0', axis=1)
master_df.columns
```

```python
filtered_df = master_df.drop_duplicates(subset=['Title', 'Description', 'Company'], keep='first', inplace=False)
print(master_df.shape)
print(filtered_df.shape)
filtered_df.to_excel("_output/desa_linked_df.xlsx")
```

```python
import ast
from collections import Counter

# Filtrar las filas que tienen valores no nulos en la columna 'Apts'
filtered_df = filtered_df.dropna(subset=['Apts'])

# Convertir las cadenas de caracteres a listas
filtered_df['Apts'] = filtered_df['Apts'].apply(ast.literal_eval)

# Crear una lista plana de todos los elementos en la columna 'Apts'
apts_list = [apt for sublist in filtered_df['Apts'] for apt in sublist]

# Contar la frecuencia de cada elemento
apt_counts = Counter(apts_list)

# Obtener los elementos más comunes y sus recuentos
top_apts = apt_counts.most_common()

print(top_apts)

```

```python
result = pd.DataFrame(top_apts, columns=['Aptitud', 'Frecuencia'])

# Mostrar el DataFrame
result.head(40)
```
